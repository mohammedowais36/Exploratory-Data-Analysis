```python
df['survival_status'] = df['survival_status'].map({1:'Yes', 2:'No'})
sns.FacetGrid(df,hue="survival_status",height= 4)\
 .map(sns.distplot,"age")\
 . add_legend();
plt.show()
```

```python
sns.FacetGrid(df,hue="survival_status",height= 4)\
 .map(sns.distplot,"operation_Year")\
 . add_legend();
plt.show()
```

```python
sns.FacetGrid(df,hue="survival_status",height= 4)\
 .map(sns.distplot,"axil_nodes")\
 . add_legend();
plt.show()
```

Observations:


1. The plot shows that those between the ages of 30 and 40 are more likely to survive, whereas those between the ages of 40 and 60 are less likely to survive. Patients between the ages of 60 and 75 have an equal probability of surviving.
2. The plot shows that survival chances are irrespective of a patient’s age and the survival chances of a patient cannot be decided by considering the age parameter only.
3. The plot shows that there were more failed procedures in the years 1960 and 1965.
4. The plot shows that the patients with no or only one node have a better chance of surviving. If there are 25 or more nodes, the survival rate is very low.


```python
Y=df.loc[df['survival_status']=='Yes']
N=df.loc[df['survival_status']=='No']

count,bin_edges=np.histogram(Y['axil_nodes'],bins=10,density=True)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'g',label="PDF FOR SURVIVAL STATUS YES")
plt.plot(bin_edges[1:],cdf,'b',label="CDF FOR SURVUVAL STATUS YES")
count,bin_edges=np.histogram(N['axil_nodes'],bins=10,density=True)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'r',label="PDF FOR SURVIVAL STATUS NO")
plt.plot(bin_edges[1:],cdf,'y',label="CDF FOR SURVIVAL STATUS NO")
plt.xlabel("NUMBER OF AXILLARY NODES")
plt.title("pdf and cdf for number of axillary nodes for both survival status")
plt.legend()
plt.plot()
```

```python
sns.boxplot(x='survival_status',y='age',data=df)
plt.show()
sns.boxplot(x='survival_status',y='operation_Year',data=df)
plt.show()
sns.boxplot(x='survival_status',y='axil_nodes',data=df)
plt.show()
```

```python
sns.violinplot(x="survival_status",y="age",data = df,height = 10)
plt.show()
sns.violinplot(x="survival_status",y="operation_Year",data = df,height = 10)
plt.show()
sns.violinplot(x="survival_status",y="axil_nodes",data = df,height = 10)
plt.show()
```

Observations:
1. The plot shows that the patients who have more than one node are unlikely to live. The greater the number of nodes, the lower the odds of survival. But there are also very few cases where the patients who had no nodes but still died within 5 years of operation.
2. The plot shows that in 1965, a higher percentage of patients who were operated on did not survive for longer than 5 years.
3. The plot shows that comparatively, more people survived till year 1960.
4. The plot shows that the patients aged between 45 and 65 died at a higher rate than the rest.
5. The plot shows that below 30 years, patients are likely to survive more.
